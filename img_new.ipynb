```python
import matplotlib.pyplot as plt

# Create a figure and display the image
fig, ax = plt.subplots()
ax.imshow(img)

# Define the line (x1, y1) to (x2, y2)
xa, ya = 0, 200  # Replace with the starting coordinates of your line
xb, yb = 200, 472  # Replace with the ending coordinates of your line

# Plot the line
line = plt.Line2D([xa, xb], [ya, yb], color='red', linewidth=1)
ax.add_line(line)

plt.show()
```

```python
m = (yb-ya)/(xb-xa)
```

```python
m1 = -1/m
```

```python
arr1 = []
for x in range(250):
  arr1.append(x)
arr1 = np.array(arr1)
```

```python
y1 = m1*(arr1 - 0) + (200 + m*0)
```

```python
arr2 = []
for x in range(50, 300):
  arr2.append(x)
arr2 = np.array(arr2)
```

```python
y2 = m1*(arr2 - 200/4) + (200 + m*200/4)
```

```python
arr3 = []
for x in range(100, 350):
  arr3.append(x)
arr3 = np.array(arr3)
```

```python
y3 = m1*(arr3 - (2/4)*200) + (200 + m*(2/4)*200)
```

```python
arr4 = []
for x in range(150, 400):
  arr4.append(x)
arr4 = np.array(arr4)
```

```python
y4 = m1*(arr4 - (3/4)*200) + (200 + m*(3/4)*200)
```

```python
arr5 = []
for x in range(200, 450):
  arr5.append(x)
arr5 = np.array(arr5)
```

```python
y5 = m1*(arr5 - (4/4)*200) + (200 + m*(4/4)*200)
```
